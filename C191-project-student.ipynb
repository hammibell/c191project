```python
#Task 2
#Get top 5 countries who has the least number of cases and plot a bar graph

import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = dataframe.dropna()

cases = df.sort_values(by = ("total_cases"))
cases

least_cases = cases.head(5)

love = least_cases["country"]
time = least_cases["total_cases"]

label = love
value = time

plt.xlabel("Top 5 countries")
plt.ylabel("Total Cases")
plt.xticks(rotation = "vertical")

plt.bar(label, value, width = 0.5, color = ("red", "orange", "yellow", "green", "blue", "purple"))

plt.show()
```

```python
#Task 4
#Get top 5 countries who has the maximum number of deaths and plot a bar graph

deaths = df.sort_values(by = "total_deaths")
deaths

sorted_deaths = deaths.tail(5)

x = sorted_deaths["total_deaths"]
y = sorted_deaths["country"]

label = y
value = x

plt.xlabel("Top 5 Countries")
plt.ylabel("Total Number of Cases")
plt.xticks(rotation = "vertical")

plt.bar(label, value, width = 0.5, color = ("red", "orange", "yellow", "green", "blue", "purple"))
plt.show()



```

```python
#Task 6
#Get top 5 countries who has the maximum number of active cases and plot a bar graph

active_cases = df.sort_values(by = "active_cases")

sorted_active = active_cases.tail(5)

country = sorted_active["country"]
active_case = sorted_active["active_cases"]

label = country
value = active_case

plt.xlabel("Top 5 Countries")
plt.ylabel("Max. Num of Active Cases")
plt.xticks(rotation = "vertical")

plt.bar(label, value, width = 0.5, color = ("red", "orange", "yellow", "green", "blue", "purple"))
plt.show()
```
